## Data Scaling and Visualization : Before and After

```python
data
```

```python
numeric_cols = data.drop("Outcome", axis=1)
```

```python
numeric_cols
```

```python
data.columns
```

```python
fig,axes=plt.subplots(2,4,figsize=(10,4))
sns.histplot(data=numeric_cols,x="Pregnancies",kde=True,ax=axes[0,0])
axes[0,0].set_title("Pregnancies")
sns.histplot(data=numeric_cols,x="Glucose",kde=True,ax=axes[0,1])
axes[0,1].set_title("Glucose")
sns.histplot(data=numeric_cols,x="BloodPressure",kde=True,ax=axes[0,2])
axes[0,2].set_title("BloodPressure")
sns.histplot(data=numeric_cols,x="SkinThickness",kde=True,ax=axes[0,3])
axes[0,3].set_title("SkinThickness")
sns.histplot(data=numeric_cols,x="Insulin",kde=True,ax=axes[1,0])

sns.histplot(data=numeric_cols,x="BMI",kde=True,ax=axes[1,1])

sns.histplot(data=numeric_cols,x="DiabetesPedigreeFunction",kde=True,ax=axes[1,2])

sns.histplot(data=numeric_cols,x="Age",kde=True,ax=axes[1,3])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax=MinMaxScaler().set_output(transform="pandas")
```

```python
minmax_scaled_df=minmax.fit_transform(numeric_cols)
```

```python
minmax_scaled_df
```

```python
# after minmax scaler
fig,axes=plt.subplots(2,4,figsize=(10,4))
sns.histplot(data=minmax_scaled_df,x="Pregnancies",kde=True,ax=axes[0,0])
axes[0,0].set_title("Pregnancies")
sns.histplot(data=minmax_scaled_df,x="Glucose",kde=True,ax=axes[0,1])
axes[0,1].set_title("Glucose")
sns.histplot(data=minmax_scaled_df,x="BloodPressure",kde=True,ax=axes[0,2])
axes[0,2].set_title("BloodPressure")
sns.histplot(data=minmax_scaled_df,x="SkinThickness",kde=True,ax=axes[0,3])
axes[0,3].set_title("SkinThickness")
sns.histplot(data=minmax_scaled_df,x="Insulin",kde=True,ax=axes[1,0])

sns.histplot(data=minmax_scaled_df,x="BMI",kde=True,ax=axes[1,1])

sns.histplot(data=minmax_scaled_df,x="DiabetesPedigreeFunction",kde=True,ax=axes[1,2])

sns.histplot(data=minmax_scaled_df,x="Age",kde=True,ax=axes[1,3])
```

### With MinMax scaling, the distribution and shape of the curve remain unchanged. The only difference is that all feature values are rescaled to fall within the range of 0 to 1.

```python
fig,axes=plt.subplots(2,4,figsize=(10,4))
sns.histplot(data=numeric_cols,x="Pregnancies",kde=True,ax=axes[0,0])
axes[0,0].set_title("Pregnancies")
sns.histplot(data=numeric_cols,x="Glucose",kde=True,ax=axes[0,1])
axes[0,1].set_title("Glucose")
sns.histplot(data=numeric_cols,x="BloodPressure",kde=True,ax=axes[0,2])
axes[0,2].set_title("BloodPressure")
sns.histplot(data=numeric_cols,x="SkinThickness",kde=True,ax=axes[0,3])
axes[0,3].set_title("SkinThickness")
sns.histplot(data=numeric_cols,x="Insulin",kde=True,ax=axes[1,0])

sns.histplot(data=numeric_cols,x="BMI",kde=True,ax=axes[1,1])

sns.histplot(data=numeric_cols,x="DiabetesPedigreeFunction",kde=True,ax=axes[1,2])

sns.histplot(data=numeric_cols,x="Age",kde=True,ax=axes[1,3])
```

```python
# standard scaler
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler().set_output(transform="pandas")
```

```python
standard_scaled_df=sc.fit_transform(numeric_cols)
```

```python
standard_scaled_df
```

```python
fig,axes=plt.subplots(2,4,figsize=(10,4))
sns.histplot(data=standard_scaled_df,x="Pregnancies",kde=True,ax=axes[0,0])
axes[0,0].set_title("Pregnancies")
sns.histplot(data=standard_scaled_df,x="Glucose",kde=True,ax=axes[0,1])
axes[0,1].set_title("Glucose")
sns.histplot(data=standard_scaled_df,x="BloodPressure",kde=True,ax=axes[0,2])
axes[0,2].set_title("BloodPressure")
sns.histplot(data=standard_scaled_df,x="SkinThickness",kde=True,ax=axes[0,3])
axes[0,3].set_title("SkinThickness")
sns.histplot(data=standard_scaled_df,x="Insulin",kde=True,ax=axes[1,0])

sns.histplot(data=standard_scaled_df,x="BMI",kde=True,ax=axes[1,1])

sns.histplot(data=standard_scaled_df,x="DiabetesPedigreeFunction",kde=True,ax=axes[1,2])

sns.histplot(data=standard_scaled_df,x="Age",kde=True,ax=axes[1,3])
```

### The shape of the distribution does not change; only the scale of the values does. Values below 0 become negative and those above 0 become positive. StandardScaler also modifies the mean and variance of the data.

```python
fig,axes=plt.subplots(2,4,figsize=(10,4))
sns.histplot(data=numeric_cols,x="Pregnancies",kde=True,ax=axes[0,0])
axes[0,0].set_title("Pregnancies")
sns.histplot(data=numeric_cols,x="Glucose",kde=True,ax=axes[0,1])
axes[0,1].set_title("Glucose")
sns.histplot(data=numeric_cols,x="BloodPressure",kde=True,ax=axes[0,2])
axes[0,2].set_title("BloodPressure")
sns.histplot(data=numeric_cols,x="SkinThickness",kde=True,ax=axes[0,3])
axes[0,3].set_title("SkinThickness")
sns.histplot(data=numeric_cols,x="Insulin",kde=True,ax=axes[1,0])

sns.histplot(data=numeric_cols,x="BMI",kde=True,ax=axes[1,1])

sns.histplot(data=numeric_cols,x="DiabetesPedigreeFunction",kde=True,ax=axes[1,2])

sns.histplot(data=numeric_cols,x="Age",kde=True,ax=axes[1,3])
```
